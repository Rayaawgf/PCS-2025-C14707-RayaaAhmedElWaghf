```python
def ex_3_1():
    """Exercise 1: List operations."""
print("\nEx 3.1:")
L = [1, 2]
L3 = 3*L
print("1. L3 =", L3)

print("2. Predictions:")
print("L3[0] =", L3[0])
print("L3[-1] =", L3[-1])
try:
    print("L3[10] =", L3[10])
except IndexError:
    print("L3[10] causes an IndexError")

L4 = [k**2 for k in L3]
print("3. L4 =", L4)

L5 = L3 + L4
print("4. L5 =", L5)
```

```python
def ex_3_5():
    """Exercise 5: Recurrence formula."""
print("\nEx 3.5: Approximate solution of u'=a*u")
h = 0.001
a = -0.5
steps = 1000

u = [math.exp(n*a*h) for n in range(3)]

for n in range(steps - 2):
    next_u = u[n+2] + h*a*((23/12)*u[n+2] - (4/3)*u[n+1] + (5/12)*u[n])
    u.append(next_u)

td = [n*h for n in range(steps + 1)]
exact = [math.exp(a*t) for t in td]
diff = [abs(exact[i] - u[i]) for i in range(steps + 1)]

print("u[0:5] =", u[:5])
print("u[-5:] =", u[-5:])
print("Max error:", max(diff))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(td, u, label='Approximation')
plt.plot(td, exact, '--', label='Exact')
plt.title("Approximate vs Exact Solution")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(td, diff)
plt.title("Absolute Error")

plt.tight_layout()
plt.show()
```

```python
def ex_3_6():
    """Exercise 6: Symmetric difference."""
def symmetric_difference(A, B):
    return (A - B) | (B - A)

A = {1, 2, 3}
B = {3, 4, 5}
print("\nEx 3.6:")
print("Custom symmetric difference:", symmetric_difference(A, B))
print("Python built-in symmetric difference:", A.symmetric_difference(B))

```

```python
def ex_3_8():
    """Exercise 8: Set operations."""
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("\nEx 3.8:")
C = A.intersection(B)
print("A.intersection(B) returns:", C)
print("A remains unchanged:", A)

A.intersection_update(B)
print("After A.intersection_update(B):", A)
```
